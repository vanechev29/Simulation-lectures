```python
import numpy as np
import matplotlib.pyplot as plt

bdata = []
bdata.append(160000000)
cuota = 1000000
for i in range(240):
    bdata.append(1.01*bdata[i]-cuota)
print(bdata[73])

i = np.arange(241)

plt.plot(i,bdata)
plt.xlabel("Meses")
plt.ylabel("Saldo")
plt.title("Saldo en el tiempo")



```

```python
#Aproximando el cambio

#vamos a graficar biomasa contra cambio

biomasa=[9.6,18.3,29,47.2,71.1,119.1,174.6,257.3]
cambio= np.diff(biomasa)

plt.plot(biomasa[0:7],cambio,'*')
plt.xlabel("Biomasa")
plt.ylabel("Cambio")

#calcule la pendiente
#toma dos puntos

pendiente=cambio[6]/biomasa[6]
x=[0,180]
y=[0,pendiente*180]
print(pendiente)

plt.plot(x,y,label="proporcionalidad")
```

```python
#hacer la solución numérica


data=[]

data.append(9.6)

for i in range(18):
    data.append((pendiente+1)*data[i])
print(data[18])

i = np.arange(19)
plt.plot(i[0:8],biomasa,'*')
plt.plot(i,data)
plt.xlabel("tiempo")
plt.ylabel("biomasa")
plt.title("Biomasa en el tiempo")
```
